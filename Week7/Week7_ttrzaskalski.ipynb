### Using pandas to import a .CSV and normalizing data + performing calculations

```python
# Below I imported the necessary modules and files

import pandas as pd
from pandas import read_csv

filename = r"C:\Users\thety\Google Drive\CUNY\IS362\\Week 7\film_ratings.csv"

# Below I created the dataframe film_ratings

film_ratings = pd.read_csv(filename)

film_ratings
```

```python
# Below I calculate the average rating by username

film_ratings.groupby('User_Name', as_index=False)['Rating'].mean()
```

```python
# Below I calculate the average rating by movie

film_ratings.groupby('Film_Name', as_index=False)['Rating'].mean()
```

```python
# Below I normalize the data by eliminating duplicate instances of User_Name and Film_Name

film_ratings2 = film_ratings.pivot(index='User_Name', columns='Film_Name', values='Rating')

film_ratings2
```

```python
# Below I calculate the average rating by username

film_ratings2.mean(axis = 1, skipna = True)
```

```python
# Below I calculate the average rating by movie

film_ratings2.mean(axis = 0, skipna = True)
```
